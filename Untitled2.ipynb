```python
# ==============================================================================
# 1. SETUP AND INSTALLATIONS
# ==============================================================================
import warnings
warnings.filterwarnings("ignore")

# Install required packages quietly
!pip install -q ddgs google-generativeai trafilatura networkx fpdf "requests[security]"

import os
import time
import requests
from urllib.parse import urlparse
import google.generativeai as genai
from ddgs import DDGS
import trafilatura
import networkx as nx
import matplotlib.pyplot as plt
from fpdf import FPDF
from google.colab import userdata

print("‚úÖ Packages installed and modules imported successfully.")

# ==============================================================================
# 2. API CONFIGURATION (SECURE METHOD)
# ==============================================================================
# IMPORTANT: Use Colab's secret manager to store your API key.
# Click the key icon on the left sidebar, add a new secret named "GEMINI_API_KEY",
# and paste your key there.
try:
    GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
    os.environ['GEMINI_API_KEY'] = GEMINI_API_KEY
    genai.configure(api_key=os.environ['GEMINI_API_KEY'])
    print("‚úÖ Gemini API key configured successfully from Colab secrets.")
except Exception as e:
    print("üõë Error configuring API key. Please ensure you have stored it as 'GEMINI_API_KEY' in Colab secrets.")
    print(f"Error details: {e}")


# ==============================================================================
# 3. MODEL INITIALIZATION
# ==============================================================================
try:
    model = genai.GenerativeModel("gemini-1.5-flash")
    print(f"‚úÖ Successfully loaded model: gemini-1.5-flash")
except Exception as e:
    print(f"üõë Error loading model: {e}")
    # Fallback to a different model if the primary one fails
    try:
        model = genai.GenerativeModel("gemini-pro")
        print("Falling back and successfully loaded model: gemini-pro")
    except Exception as e2:
        print(f"üõë Could not load any Gemini model. Error: {e2}")


# ==============================================================================
# 4. CORE RESEARCH AGENT FUNCTIONS
#
# This section defines the functions for the multi-step research process:
# - Planner: Breaks down the main query into sub-queries.
# - Searcher: Finds relevant online sources for each sub-query.
# - Reader: Fetches and extracts content from source URLs.
# - Verifier: Extracts factual claims from the content.
# - Briefer: Synthesizes claims into a final report.
# ==============================================================================

def planner(query):
    """
    Generates specific, targeted sub-queries from a broad research question.
    """
    try:
        response = model.generate_content(
            f"""As a world-class research assistant, break this primary research question into 4-5 specific, answerable sub-queries. These sub-queries should be focused and distinct to guide a comprehensive investigation.

            **Primary Question:** "{query}"

            Provide only the sub-queries, one per line, without any numbering, bullet points, or introductory text."""
        )
        subqs = [line.strip() for line in response.text.split("\n") if line.strip() and len(line) > 10]
        return subqs if subqs else [query]  # Fallback to the main query
    except Exception as e:
        print(f"‚ö†Ô∏è Planner Error: {e}. Using default sub-queries.")
        return [
            f"health benefits of {query}",
            f"risks and side effects of {query}",
            f"impact of {query} on metabolic health",
            f"long-term effects of {query}"
        ]

def searcher(main_query, sub_query, max_results=4):
    """
    Searches the web for a sub-query, prioritizing reliable domains.
    """
    results = []
    # Combine queries for better search context
    search_term = f"{main_query} {sub_query}"
    print(f"  -> Searching for: \"{search_term}\"")
    try:
        with DDGS() as ddgs:
            # Fetch more results than needed to allow for filtering
            search_results = list(ddgs.text(search_term, max_results=max_results * 2))

            # Domains to prioritize (will be moved to the top)
            priority_domains = ['nih.gov', 'ncbi.nlm.nih.gov', 'mayoclinic', 'clevelandclinic', 'health.harvard.edu', 'hopkinsmedicine', 'webmd', 'healthline', 'medicalnewstoday']
            # Domains to exclude
            excluded_domains = ['quora', 'pinterest', 'facebook', 'twitter', 'youtube', 'tiktok', 'reddit']

            temp_results = []
            for r in search_results:
                if not r.get("href"):
                    continue

                domain = urlparse(r["href"]).netloc.lower()
                if any(bad in domain for bad in excluded_domains):
                    continue

                # Store title, URL, and priority status
                is_priority = any(good in domain for good in priority_domains)
                temp_results.append({"title": r.get("title", "No title"), "url": r["href"], "priority": is_priority})

            # Sort results to put priority domains first
            sorted_results = sorted(temp_results, key=lambda x: x['priority'], reverse=True)

            # Return only the title and URL of the top results
            results = [(item['title'], item['url']) for item in sorted_results[:max_results]]

    except Exception as e:
        print(f"  -> ‚ö†Ô∏è Search Error for '{sub_query}': {e}")
    return results

def reader(url):
    """
    Reads and extracts the main textual content from a URL using requests and trafilatura.
    """
    try:
        # Use requests to fetch the page with a User-Agent and timeout
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers, timeout=15)
        response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)

        # Pass the downloaded HTML to trafilatura to extract the main content
        content = trafilatura.extract(response.text)
        return content if content and len(content) > 200 else None
    except requests.exceptions.RequestException as e:
        print(f"  -> ‚ö†Ô∏è Reader Error (Request): Could not fetch {url}. Reason: {e}")
        return None
    except Exception as e:
        print(f"  -> ‚ö†Ô∏è Reader Error (Extraction): Could not process {url}. Reason: {e}")
        return None

def verifier(docs, query):
    """
    Extracts factual claims from a list of documents using the Gemini model.
    """
    claims = []
    for idx, text in enumerate(docs):
        if text and len(text) > 200:
            try:
                # Use a larger portion of the text for better context
                truncated_text = text[:20000]

                prompt = f"""From the text provided below, extract up to 5 key factual claims that directly address the research question: "{query}".

                - Each claim must be a concise, verifiable statement.
                - Ignore advertisements, navigation links, and boilerplate text.
                - For each claim, determine the reliability of the source on a scale from 1 (unreliable) to 5 (highly reliable, e.g., scientific journal, government health site).

                **Format your output exactly as follows, with each claim on a new line:**
                [Source {idx}] [Reliability Score] Claim text.

                **Text to Analyze:**
                {truncated_text}
                """

                resp = model.generate_content(prompt)
                if resp.text:
                    claims.append(resp.text)
                time.sleep(1)  # Rate limiting
            except Exception as e:
                print(f"  -> ‚ö†Ô∏è Verification Error for document {idx}: {e}")
    return "\n".join(claims)

def briefer(claims, query, sources):
    """
    Generates a structured research brief from a set of verified claims.
    """
    if not claims:
        return "No valid claims were extracted. The research could not be completed."

    source_list = "\n".join([f"[{i}] {url}" for i, (_, url) in enumerate(sources)])

    prompt = f"""You are a professional research analyst. Synthesize the following verified claims into a comprehensive and well-structured research brief about the query: "{query}".

    **Verified Claims:**
    {claims}

    **Instructions:**
    1.  **Introduction:** Start with a brief overview of the topic.
    2.  **Key Findings:** Group related claims into thematic paragraphs. Use a neutral, objective tone.
    3.  **Cite Your Sources:** At the end of each sentence or major point, cite the source using the format `[Source X]`.
    4.  **Potential Risks/Contradictions:** If the claims present conflicting information or risks, highlight them in a dedicated section.
    5.  **Conclusion:** Provide a balanced summary of the overall findings.
    6.  **References:** Do NOT create a reference list in the brief itself; one will be appended automatically.

    Produce a high-quality, readable brief.
    """
    try:
        resp = model.generate_content(prompt)
        final_brief = resp.text + "\n\n" + "="*20 + "\nSOURCES REFERENCED\n" + "="*20 + "\n" + source_list
        return final_brief
    except Exception as e:
        return f"üõë Error generating brief: {str(e)}"

# ==============================================================================
# 5. UTILITY FUNCTIONS (GRAPHING AND EXPORTING)
# ==============================================================================

def build_citation_graph(sources):
    """
    Creates and displays a graph of the research sources.
    """
    if not sources:
        print("No sources to visualize.")
        return

    G = nx.Graph()
    labels = {}
    for i, (_, url) in enumerate(sources):
        domain = urlparse(url).netloc.replace("www.", "")
        node_id = f"Source {i}"
        G.add_node(node_id)
        labels[node_id] = domain

    if not G.nodes:
        print("Graph could not be created as no nodes were available.")
        return

    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=False, node_color="skyblue", node_size=2000, edge_color="gray")
    nx.draw_networkx_labels(G, pos, labels, font_size=10, font_weight="bold")
    plt.title("Citation Network of Research Sources", size=16)
    plt.show()

def export_pdf(text, filename="research_brief.pdf"):
    """
    Exports the research brief text to a PDF file.
    """
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=11)
    # Encode text to handle special characters gracefully
    lines = text.split('\n')
    for line in lines:
        try:
            pdf.multi_cell(0, 10, line.encode('latin-1', 'replace').decode('latin-1'))
        except Exception as e:
            print(f"Warning: Could not add line to PDF: {line[:50]}... Error: {e}")

    try:
        pdf.output(filename)
        print(f"‚úÖ Research brief successfully exported to {filename}")
    except Exception as e:
        print(f"üõë Error exporting PDF: {e}")

# ==============================================================================
# 6. MAIN EXECUTION ORCHESTRATOR
# ==============================================================================

def conduct_research(query):
    """
    Orchestrates the entire research process from planning to exporting.
    """
    print(f"üî¨ Starting new research project for: \"{query}\"\n")

    # Step 1: Plan - Break down the query
    print("Step 1: Planning sub-queries...")
    sub_queries = planner(query)
    print(f"Generated {len(sub_queries)} sub-queries:\n- " + "\n- ".join(sub_queries) + "\n")

    # Step 2: Search - Gather sources for each sub-query
    print("Step 2: Searching for sources...")
    all_urls = []
    for sq in sub_queries:
        results = searcher(query, sq)
        all_urls.extend(results)
        time.sleep(1) # Be respectful to the search API

    # Remove duplicate URLs while preserving order
    unique_urls = list(dict.fromkeys(all_urls))
    print(f"Found {len(unique_urls)} unique sources.\n")

    # Step 3: Read - Extract content from sources
    print("Step 3: Reading and extracting content from sources...")
    docs = []
    source_map = [] # Keep track of which doc corresponds to which URL
    for i, (title, url) in enumerate(unique_urls):
        print(f"  ({i+1}/{len(unique_urls)}) Reading: {title[:60]}...")
        content = reader(url)
        if content:
            docs.append(content)
            source_map.append((title, url))
        time.sleep(0.5)
    print(f"Successfully read and processed {len(docs)} documents.\n")

    if not docs:
        print("üõë No documents could be read. Aborting research.")
        return

    # Step 4: Verify - Extract factual claims
    print("Step 4: Verifying sources and extracting claims...")
    claims = verifier(docs, query)
    print(f"Extracted claims from documents.\n")

    # Step 5: Brief - Generate the final report
    print("Step 5: Synthesizing claims into a research brief...")
    brief = briefer(claims, query, source_map)
    print("\n" + "="*50 + "\nüìÑ FINAL RESEARCH BRIEF\n" + "="*50)
    print(brief)
    print("="*50 + "\n")

    # Step 6: Graph - Visualize the sources
    print("Step 6: Building citation graph...")
    build_citation_graph(source_map)
    print("\n")

    # Step 7: Export - Save the report as a PDF
    print("Step 7: Exporting brief to PDF...")
    filename = f"{query.replace(' ', '_').lower()}_research_brief.pdf"
    export_pdf(brief, filename)

    print("\n‚úÖ Research project completed!")
    return brief

# ==============================================================================
# 7. EXAMPLE USAGE
# ==============================================================================
if __name__ == "__main__":
    research_query = "What are the health impacts of intermittent fasting?"
    final_report = conduct_research(research_query)
```
